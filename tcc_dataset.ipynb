# Execução da ferramenta Designite

A ferramenta DesigniteJava é usada para detectar automaticamente code smells de diferentes categorias: Design Smells, Implementation Smells, Architecture Smells e Test Smells.

Este script automatiza o processo de análise de projetos Java utilizando a ferramenta DesigniteJava. Inicialmente, o script localiza projetos contendo código Java, executa a ferramenta para detectar code smells e gerar relatórios em formato CSV. Em seguida, todos os arquivos DesignSmells.csv gerados são coletados e consolidados em um único dataset, que será utilizado nas etapas posteriores de análise e aplicação de algoritmos de aprendizado de máquina.

```python
import subprocess
import glob
import os
import pandas as pd

# CONFIGURAÇÃO DE CAMINHOS

designite_path = "/content/designite/DesigniteJava.jar"
base_path = "/content/ck/repos_java"

print("Iniciando analise de projetos Java\n")

# LOCALIZAR DIRETÓRIOS COM CÓDIGO JAVA

java_paths = glob.glob(f"{base_path}/**/src/main/java", recursive=True)

print(f"Projetos Java encontrados: {len(java_paths)}\n")

# EXECUTAR DESIGNITE EM CADA PROJETO

for path in java_paths:

    project_dir = os.path.abspath(os.path.join(path, "../../.."))
    output_dir = os.path.join(project_dir, "designite_output")

    os.makedirs(output_dir, exist_ok=True)

    print(f"Analisando projeto: {project_dir}")

    try:

        subprocess.run(
            ["java", "-jar", designite_path, "-i", project_dir, "-o", output_dir],
            check=True
        )

        print("Análise concluída")

        type_metrics_file = os.path.join(output_dir, "TypeMetrics.csv")

        if os.path.exists(type_metrics_file):
            with open(type_metrics_file, "r") as f:
                class_count = sum(1 for _ in f) - 1
            print(f"Classes analisadas: {class_count}")

        java_files = glob.glob(f"{project_dir}/**/*.java", recursive=True)
        print(f"Arquivos Java encontrados: {len(java_files)}\n")

    except subprocess.CalledProcessError as e:
        print(f"Erro ao analisar {project_dir}: {e}\n")


# COLETAR ARQUIVOS DESIGNSMELLS

print("\nColetando arquivos de code smells...")

files = glob.glob(f"{base_path}/**/designite_output/DesignSmells.csv", recursive=True)

print(f"Arquivos DesignSmells encontrados: {len(files)}")

df_list = []

# LER E CONSOLIDAR OS DATASETS

for f in files:

    try:

        df = pd.read_csv(f)

        if df.shape[0] > 0:

            project_name = f.split("/")[-3]

            df["Project"] = project_name

            df_list.append(df)

    except Exception as e:

        print("Erro lendo arquivo:", f)
        print(e)


# GERAR DATASET CODE SMELLS

if len(df_list) > 0:

    dataset = pd.concat(df_list, ignore_index=True)

    print("\nDataset final gerado")
    print("Número de registros:", dataset.shape[0])
    print("Número de colunas:", dataset.shape[1])

    output_file = "/content/dataset_code_smells.csv"

    dataset.to_csv(output_file, index=False)

    print(f"\nDataset salvo em: {output_file}")

else:

    print("Nenhum code smell encontrado nos projetos analisados.")
```

# Geração de rótulos automáticos

Como nem todas as classes possuem anotações manuais de smells, é feita uma rotulação heurística.
Classes com valores acima da média para métricas como WMC, CBO, LCOM ou RFC são marcadas como code_smell = 1.

```python
import pandas as pd
import glob
import os

base_path = "/content/ck/repos_java"

output_files = glob.glob(f"{base_path}/**/class.csv", recursive=True)
print(f"{len(output_files)} arquivos CSV encontrados para consolidação.\n")

if not output_files:
    raise FileNotFoundError("Nenhum arquivo de métricas CK encontrado nas pastas de saída.")

frames = []

for f in output_files:
    try:
        df_part = pd.read_csv(f)
        df_part["source_file"] = os.path.basename(f)  # rastreabilidade
        df_part["project"] = f.split("/")[3] if len(f.split("/")) > 3 else "unknown"
        frames.append(df_part)
        print(f"Lido: {f} ({df_part.shape[0]} linhas)")
    except Exception as e:
        print(f"Erro ao ler {f}: {e}")

if frames:
    df = pd.concat(frames, ignore_index=True)
    print(f"\n dataset combinado com {df.shape[0]} linhas e {df.shape[1]} colunas.")

    if {"wmc", "cbo", "lcom", "rfc"}.issubset(df.columns):
        limiar_wmc = df["wmc"].mean()
        limiar_cbo = df["cbo"].mean()
        limiar_lcom = df["lcom"].mean()
        limiar_rfc = df["rfc"].mean()

        df["code_smell"] = (
            (df["wmc"] > limiar_wmc) |
            (df["cbo"] > limiar_cbo) |
            (df["lcom"] > limiar_lcom) |
            (df["rfc"] > limiar_rfc)
        ).astype(int)

        print(f"Rótulo 'code_smell' criado com base em múltiplas métricas (WMC, CBO, LCOM, RFC)")
    else:
        df["code_smell"] = 0
        print("Métricas principais ausentes. Criada coluna 'code_smell' dummy = 0.")

    output_path = "/content/dataset_metrics.csv"
    df.to_csv(output_path, index=False)
    print(f"\nDataset consolidado salvo em: {output_path}")
else:
    print("Nenhum arquivo CSV válido foi processado.")
```

# Consolidação dos datasets

Os arquivos gerados pelas ferramentas CK e Designite são integrados, formando o dataset final dataset_final.csv.
A junção é realizada com base nos identificadores de arquivo e classe, combinando métricas estruturais (features) e rótulos de smells (labels).
Esse dataset é a principal entrada para os modelos de aprendizado de máquina.

```python
import pandas as pd

# CARREGAR DATASETS

df_metrics = pd.read_csv("/content/dataset_metrics.csv")
df_smells = pd.read_csv("/content/dataset_code_smells.csv")

print("Datasets carregados com sucesso")

# PADRONZAR NOME DAS CLASSES

df_metrics["class_name"] = df_metrics["class"].str.split(".").str[-1]

df_smells["class_name"] = df_smells["Class"]

# MERGE DOS DATASETS

dataset_final = pd.merge(
    df_metrics,
    df_smells,
    on="class_name",
    how="left"
)

print("Merge realizado")

# CRIAR VARIÁVEL BINÁRIA DE SMELL

dataset_final["has_smell"] = dataset_final["Smell"].notna().astype(int)

# TRANSFORMAR SMELLS AGRUPANDO POR CLASSE

smell_matrix = pd.get_dummies(dataset_final["Smell"])

dataset_final = pd.concat([dataset_final, smell_matrix], axis=1)

dataset_final = dataset_final.groupby("class_name").max().reset_index()

# TRATAR VALORES AUSENTES

dataset_final.fillna(0, inplace=True)

# REMOVER COLUNAS DESNECESSÁRIOS

dataset_final = dataset_final.drop(
    columns=[
        "Smell",
        "Package",
        "Description",
        "File",
        "Project",
        "assignmen"  # coluna vazia
    ],
    errors="ignore"
)

# SALVAR DATASET FINAL

output_path = "/content/dataset_final.csv"

dataset_final.to_csv(output_path, index=False)

print(f"Dataset final salvo em: {output_path}")
print("Shape:", dataset_final.shape)
```
